```python
import matplotlib.pyplot as plt
import seaborn as sns



plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=20, kde=True, color='skyblue')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
department_counts = df['JobTitle'].value_counts()
plt.pie(department_counts, labels=department_counts.index, autopct='%0.1f%%', startangle=90)
plt.title('Employee Distribution by Department')
plt.show()
```

```python
grouped_data = df.groupby('JobTitle')['TotalPay'].mean()
grouped_data
```

```python
grouped_data = df.groupby(['Id', 'JobTitle']).agg({
    'TotalPay': ['mean', 'median', 'min', 'max', 'std', 'count']
}).reset_index()
```

```python
grouped_data
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='JobTitle', y=('TotalPay', 'mean'), data=grouped_data, palette='viridis')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()
```

```python
correlation = df['TotalPay'].corr(df['Year'])
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year', y='TotalPay', data=df, color='blue')
plt.title('Scatter Plot: Salary vs Years')
plt.xlabel('Years')
plt.ylabel('Salary')
plt.show()
```
